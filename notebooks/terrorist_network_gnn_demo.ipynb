## 3. Quick Demo Run (Fast Testing)

Run with minimal networks for quick testing (5-10 minutes)

```python
import sys
sys.path.append("./src")

from terrorist_network_dataset import NetworkConfig, TerroristNetworkGenerator, DisruptionEvaluator
from terrorist_network_disruption import MultiLayerTemporalGNN, EnhancedCriticalNodeDetector
from training import EnhancedTemporalGNNTrainer

# Quick test configuration
network_config = NetworkConfig(
    initial_nodes=30,
    max_nodes=50,
    recruitment_rate=0.05,
    dropout_rate=0.02,
    physical_density=0.15,
    digital_density=0.35,
    financial_density=0.08,
    ideological_density=0.30,
    operational_density=0.06
)

model_config = {
    'num_node_features': 64,
    'num_edge_features': 32,
    'hidden_dim': 128
}

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Generate networks
print("\nGenerating test networks...")
generator = TerroristNetworkGenerator(network_config, seed=42)
networks = []
for i in range(3):
    network = generator.generate_temporal_network(
        num_timesteps=10,
        num_node_features=model_config['num_node_features'],
        num_edge_features=model_config['num_edge_features'],
        device=device
    )
    networks.append(network)
    print(f"  Network {i+1}/3 generated")

# Create model
print("\nCreating Temporal GNN Model...")
model = MultiLayerTemporalGNN(
    num_node_features=model_config['num_node_features'],
    num_edge_features=model_config['num_edge_features'],
    hidden_dim=model_config['hidden_dim'],
    num_layers=5,
    layer_fusion='attention',
    num_temporal_layers=3,
    num_graph_layers=3,
    num_attention_heads=8,
    memory_size=100
).to(device)

print(f"‚úì Model created with {sum(p.numel() for p in model.parameters()):,} parameters")

# Train model
print("\nTraining Model...")
trainer = EnhancedTemporalGNNTrainer(
    model,
    learning_rate=1e-3,
    weight_decay=1e-5,
    device=device,
    negative_sampling_strategy='hard_negative'
)

history = trainer.fit(
    networks[:2],
    networks[2:],
    num_epochs=30,
    early_stopping_patience=10,
    verbose=True
)

print("\n‚úì Quick demo complete!")
```

## 5. View Results

```python
# Load and display results
import json
from IPython.display import Image, display
import glob
import numpy as np

# Choose which experiment to view
experiment_dir = 'experiments/results'

# Load results JSON
with open(f'{experiment_dir}/results.json', 'r') as f:
    results = json.load(f)

print("="*80)
print("EXPERIMENTAL RESULTS SUMMARY")
print("="*80)

# Training history
if 'training_history' in results:
    history = results['training_history']
    print(f"\nTraining:")
    print(f"  Final Train Loss: {history['train_loss'][-1]:.4f}")
    print(f"  Final Val Loss: {history['val_loss'][-1]:.4f}")
    print(f"  Best Val Loss: {min(history['val_loss']):.4f}")

# Baseline comparison
if 'baseline_comparison' in results:
    print("\nBaseline Comparison:")
    baseline_means = {k: sum(v)/len(v) for k, v in results['baseline_comparison'].items()}
    for method, mean in sorted(baseline_means.items(), key=lambda x: x[1], reverse=True):
        marker = " ‚òÖ" if method == 'our_method' else ""
        print(f"  {method:<25}{marker} {mean:.4f}")

# Statistical tests
if 'statistical_tests' in results and 'baseline_comparison' in results['statistical_tests']:
    print("\nStatistical Significance:")
    comparisons = results['statistical_tests']['baseline_comparison']['comparisons']
    for method, stats in comparisons.items():
        if 'bonferroni_significant' in stats and stats['bonferroni_significant']:
            print(f"  ‚úì {method}: Significantly different (p < Bonferroni Œ±)")

# Ablation study
if 'ablation_study' in results:
    print("\nAblation Study - Component Importance:")
    ablation = results['ablation_study']
    components = [(k, v['performance_drop']) for k, v in ablation.items()
                  if k != 'full_model' and 'performance_drop' in v]
    components.sort(key=lambda x: x[1], reverse=True)
    for i, (component, drop) in enumerate(components[:5], 1):
        print(f"  {i}. {ablation[component]['name']}: {drop:.4f}")
```

## 7. Advanced: Custom Analysis

```python
# Example: Analyze specific components
import matplotlib.pyplot as plt

# Load results
with open(f'{experiment_dir}/results.json', 'r') as f:
    results = json.load(f)

# Custom analysis example: Compare Q1 strategies in detail
if 'q1_critical_nodes' in results:
    q1_results = results['q1_critical_nodes']

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Box plot
    strategies = list(q1_results.keys())
    data = [[r['disruption'] for r in q1_results[s]] for s in strategies]

    axes[0].boxplot(data, labels=strategies)
    axes[0].set_ylabel('Disruption Score')
    axes[0].set_title('Q1 Strategy Distribution')
    axes[0].tick_params(axis='x', rotation=45)
    axes[0].grid(alpha=0.3)

    # Violin plot
    parts = axes[1].violinplot(data, positions=range(len(strategies)),
                                showmeans=True, showmedians=True)
    axes[1].set_xticks(range(len(strategies)))
    axes[1].set_xticklabels(strategies, rotation=45, ha='right')
    axes[1].set_ylabel('Disruption Score')
    axes[1].set_title('Q1 Strategy Distribution (Violin)')
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.show()
```
